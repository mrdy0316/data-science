```r
# CSVファイルの読み込み
body.data <- read.csv("../data/body.csv", header=T, stringsAsFactors=F)

# 線形回帰モデルの構築
body.lm1 <- lm(weight~height, data=body.data)
summary(body.lm1)

# 回帰診断図の実行
plot(body.lm1, which=1)

# 逓減回帰モデルの構築
body.lm2 <- lm(weight~log(height), data=body.data)
summary(body.lm2)
plot(body.lm2, which=1)
```
